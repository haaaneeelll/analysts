- 분석, 머신러닝 및 데이터 과학을 다루는 대부분의 책은 엔지너어 또는 마케터와 같이 서로 다른 분야에 있는 실무자가 당면한 문제를 동일한 접근 방식과 도구로 해결할 수 있다고 암묵적으로 가정한다. 물론, 다루는 변수도 다르고 바탕이 되는 지식도 도메인별로 다르지만 K-평균 클러스터링을 터빈의 센서 데이터에 적용하든 소셜 미디어의 게시물에 적용하든 핵심이 되는 접근법은 K-평균 클러스터링으로 동일하다는 의미이다. 
- 이러한 방식으로 머신러닝 도구를 도입하면서 기업은 분석 대상의 행동 경향을 대체로 정확하게 예측할 수 있게 되었지만 실제로 어떤 분석 원리로 결과를 도출할 수 있는지에 대해 깊고 풍부하게 이해하는 것은 포기해야 했다. 결국 데이터 과학 모델을 블랙박스라고 비판하는 목소리도 생겨났다.

- 정확하지만 불투명한 예측 방법을 알아보는 대신 '무엇이 행동을 유발하는가?'라는 질문에 답을 찾아야 한다.
- 정말로 그 고객이 그 이메일을 계기로 서비스를 구독할까?
- 어떤 고객층에게 이메일을 보내야 할까?
- 나이가 많은 고객이 구매하는 상품이 따로 있을까?
- 고객 경험이 충성도와 유지율에 어떤 영향을 줄까?
> 행동을 설명하고 그 원인을 파악하는 것으로 분석의 관점을 전환하면 자신의 일에 확신을 가질 터

- 데이터를 이해해야 한다. 데이터에 대한 감각을 늘려야 한다.
